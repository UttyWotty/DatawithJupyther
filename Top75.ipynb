```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/NBA_75th_Anniversary_Team"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html.parser")

print(soup.prettify())
```

```python
table = soup.find_all('table')[1]
```

```python
headers = [header.text.strip() for header in table.find_all('tr')[0].find_all('th')]
```

```python
rows = table.find_all('tr')[1:]
data = [[cell.text.strip() for cell in row.find_all(['td', 'th'])] for row in rows]
```

```python
# Create a DataFrame
df = pd.DataFrame(data, columns=headers)

# Print the DataFrame
print(df)
```

```python
df.to_csv("NBA75.csv")
```
